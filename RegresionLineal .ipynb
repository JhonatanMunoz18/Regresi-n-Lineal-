```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Crear el modelo
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X_train, y_train)
```

```python
# Predicciones con los datos de prueba
y_pred = modelo.predict(X_test)

# Evaluación del modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Error cuadrático medio:", mse)
print("Coeficiente de determinación R^2:", r2)
```

##Error Cuadrático Medio (ECM): 36,371,600,176.79

Si los precios están en un rango de cientos de miles, este error podría ser significativo.
Coeficiente de Determinación R^2: 0.4110

Interpretación: Un R2R^2R2 de 0.41 significa que el modelo explica aproximadamente el 41.1% de la variabilidad en el precio de venta de los automóviles. Esto sugiere que, aunque hay alguna correlación entre el año, los kilómetros recorridos y el precio de venta, hay otros factores no considerados en el modelo que también influyen en el precio. Un R2R^2R2 por debajo de 0.5 generalmente indica que el modelo podría mejorarse.
